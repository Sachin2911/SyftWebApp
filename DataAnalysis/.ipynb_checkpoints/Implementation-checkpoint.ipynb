```python
import pandas as pd
pd.set_option('display.max_columns', None)
import numpy as np
```

```python
merged_df = pd.read_csv("./Data/merged.csv")
```

```python
#First we need to drop all columns with issue date nan to avoid problems when charts are displayed and then we need
# to replace all the nan values with 0
```

```python
merged_df.dropna(subset=['issue_date'], inplace=True)
```

```python
#We first need to filter based on the query fields product and dates
```

```python
query = ['profit', 'TIN001', '2021-04-05T00:00:00.000Z', '2023-08-05T00:00:00.000Z', 'arima']

#filtering after startdate
filtered_df = merged_df[merged_df["issue_date"]>=query[2]]
#filtering before enddate
filtered_df = filtered_df[merged_df["issue_date"]<=query[3]]
filtered_df = filtered_df[merged_df["item_code"]==query[1]]
```

```python
# Sorting out revenue expense and profit
```

```python
merged2_df = pd.read_csv("./Data/merged.csv")
```

```python
items = pd.read_csv("./Data/items.csv")
```

```python
items
```

```python
merged2_df["code"] = merged2_df["item_code"]
```

```python
merged = pd.merge(merged2_df, items[['code', 'sale_unit_price']], on='code', how='left')
merged = pd.merge(merged, items[['code', 'purchase_unit_price']], on='code', how='left')
merged.fillna(0, inplace=True)
```

```python
merged["revenue"] = merged["quantity"]*merged["sale_unit_price"]
merged["expense"] = merged["quantity"]*merged["purchase_unit_price"]
merged["profit"] = merged["revenue"] - merged["expense"]
```
