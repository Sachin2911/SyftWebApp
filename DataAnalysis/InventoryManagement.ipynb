```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

def stl_decomposition(data, item_code, date_col, value_col):
    # Filter data for the given item_code
    item_data = data[data['item_code'] == item_code]
    
    # Ensure date column is in datetime format
    item_data[date_col] = pd.to_datetime(item_data[date_col])
    
    # Aggregate data (assuming monthly for simplicity)
    aggregated_data = item_data.groupby(item_data[date_col].dt.to_period("M")).agg({
        value_col: 'sum'
    }).reset_index()
    
    # Convert period back to datetime for STL
    aggregated_data[date_col] = aggregated_data[date_col].dt.to_timestamp()
    
    # Perform STL decomposition
    result = sm.tsa.seasonal_decompose(item_data[date_col], model='additive', period=4)
    
    # Plot the decomposition
    result.plot()
    plt.suptitle(f'STL Decomposition for item_code: {item_code}')
    plt.tight_layout()
    plt.subplots_adjust(top=0.9)
    plt.show()
```

```python
stl_decomposition(merged, "ARI001", "issue_date", "profit")
```
